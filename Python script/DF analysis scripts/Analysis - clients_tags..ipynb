### Looping over the tags to break them down into unique keywords

#### Identifying unique tags

```python
#Creating a list of all tags combination
tags_raw = []
#for each list of tags per client
for combo in clients.tags:
    tags_raw.append(combo)

#Creating a final list of each UNIQUE tags you can find to create the dict 
tags_unique=[]
#for each groups of tags per client
for elements in tags_raw:
    if elements:
        #for each tags in a group
        for tags in elements.split(", "):
            if tags not in tags_unique:
                tags_unique.append(tags)

#Creating a cleaned list of every tags per client to count the occurences of each tags per client (removing the None value)
tags_final= []
#for each list of tags per client
for combo in tags_raw:
    l=[]
    #if combo is not null (to delete the None values)
    if combo: 
        for x in combo.split(','):
            #strip allows to remove blank spaces and harmonize the list
            l.append(x.strip())
    tags_final.append(l)

print(len(tags_final))
```
