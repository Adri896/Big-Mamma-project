Datetime and UTC transformation

```python
#Cast data to datetime
orders[['reservation_date','arrival_time','seated_time','seated_time_local','ticket_date','accounting_date','time_of_opening','time_of_op_local','time_of_close_local','time_of_close_and_paid']]=orders[['reservation_date','arrival_time','seated_time','seated_time_local','ticket_date','accounting_date','time_of_opening','time_of_op_local','time_of_close_local','time_of_close_and_paid']].apply(pd.to_datetime,format="mixed")  
```

```python
#Remove UTC
orders['seated_time']=orders['seated_time'].dt.tz_localize(None)
orders['time_of_close_and_paid']=orders['time_of_close_and_paid'].dt.tz_localize(None)
orders['time_of_opening']=orders['time_of_opening'].dt.tz_localize(None)
```

Converting bools to 0/1

```python
#Function to convert the bool to int
def bool_converter(x):
    if x==True:
        return 1
    else:
        return 0
```

```python
#Conversion of columns with a boolean value to 0/1
orders[['is_transfer','is_update','is_void','is_back_office','is_neg_sales','is_zero_sales','is_pos_sales','is_off_dir','is_full_disc']]=orders[['is_transfer','is_update','is_void','is_back_office','is_neg_sales','is_zero_sales','is_pos_sales','is_off_dir','is_full_disc']].map(bool_converter)
```
